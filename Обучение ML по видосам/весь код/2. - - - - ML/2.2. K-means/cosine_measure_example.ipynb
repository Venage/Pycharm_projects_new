# Пример сравнения текстов по косинусной мере

```python
def readVacanciesFromFile( fileName ):
    """
    Возвращает лист с текстами вакансий из файла fileName
    """
    
    with open( fileName, 'r', encoding = 'utf-8' ) as f:
        return [ line.strip() for line in f ]
```

```python
accountantTexts = readVacanciesFromFile( 'бухгалтерские вакансии.txt' )
```

```python
accountantTexts[:2]
```

```python
developerTexts = readVacanciesFromFile( 'вакансии разработчиков.txt' )
```

```python
developerTexts[:1]
```

```python
texts = accountantTexts + developerTexts
```

```python
len( texts )
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer()
```

```python
import pandas as pd
```

```python
matrix_count = count_vec.fit_transform( texts ).toarray()
words = [ x[0] for x in sorted( count_vec.vocabulary_.items(), key=lambda x: x[1] ) ]
```

```python
pd.DataFrame(matrix_count, columns=words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec = TfidfVectorizer( norm = None, smooth_idf = False )
matrix_tfidf = tfidf_vec.fit_transform( texts ).toarray()
```

```python
words = [ x[0] for x in sorted( tfidf_vec.vocabulary_.items(), key=lambda x: x[1] ) ]
pd.DataFrame( matrix_tfidf, columns=words )
```

```python
def cos_m(x, y):
    sum_ = 0
    len_x = 0
    len_y = 0
    for i in range(len(x)):
        sum_ += x[i]*y[i]
        len_x += x[i]**2
        len_y += y[i]**2
        
    return sum_/(len_x**0.5*len_y**0.5)
```

```python
X = []
Y = []

for x in range(1, 2000):
    X.append(x)
    Y.append( cos_m( matrix_tfidf[x], matrix_tfidf[0] ) )
```

```python
import matplotlib
%matplotlib inline

matplotlib.pyplot.plot(X,Y)
```
