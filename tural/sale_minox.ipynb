```python
%%time
#Importing the required modules
import pandas as pd
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter


#Creating an instance of Nominatim Class
geolocator = Nominatim(user_agent="my_request")
 
#applying the rate limiter wrapper
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
 
#Applying the method to pandas DataFrame
df_adress['location'] = df_adress['Город_short'].apply(geocode)
df_adress['Lat'] = df_adress['location'].apply(lambda x: x.latitude if x else None)
df_adress['Lon'] = df_adress['location'].apply(lambda x: x.longitude if x else None)
 
df_adress
```

```python
%%time
import folium
from folium.plugins import MarkerCluster
import random

# create a map object
mapObj = folium.Map(location=[55.735628,37.621380], zoom_start=10, control_scale=True)
df_adress_map = df_new[['Номер','Город']].drop_duplicates().merge(df_adress, how = 'left')

mCluster = MarkerCluster(name="Markers Demo").add_to(mapObj)

for i,j,z in zip(df_adress_map['Lat'],df_adress_map['Lon'],df_adress_map['Город']):
    try:
        folium.Marker(location=[i, j], popup="{2} - {0}, {1}".format(i, j, z)).add_to(mCluster)
    except:
        print()

folium.LayerControl().add_to(mapObj)

# save the map object as html
mapObj.save(r"C:\PyProjects\poject_1\tural\map.html")
```
