```python
df_final = pd.DataFrame(columns=['Категория','Категория подробная','Количество'])

month  = ['янв_20', 'фев_20', 'мар_20',
       'апр_20', 'май_20', 'июн_20', 'июл_20', 'авг_20', 'сен_20', 'окт_20',
       'ноя_20', 'дек_20', 'янв_21', 'фев_21', 'мар_21', 'апр_21', 'май_21',
       'июн_21', 'июл_21', 'авг_21', 'сен_21', 'окт_21', 'ноя_21', 'дек_21',
       'янв_22', 'фев_22', 'мар_22', 'апр_22', 'май_22', 'июн_22', 'июл_22',
       'авг_22', 'сен_22', 'окт_22']

for i in range(len(month)):
    df_kat_podcat_grouped_x = df_kat_podcat_grouped.iloc[:, [0,1,i+2]]
    df_kat_podcat_grouped_x['Месяц'] = month[i].split('_')[0]
    df_kat_podcat_grouped_x['Год'] = str(20)+month[i].split('_')[1]
    df_kat_podcat_grouped_x['Количество'] = df_kat_podcat_grouped.iloc[:, [i+2]]
    df_kat_podcat_grouped_x = df_kat_podcat_grouped_x.drop(month[i], axis=1)
    df_final = df_final.append(df_kat_podcat_grouped_x, ignore_index=True)
```
